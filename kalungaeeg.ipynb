```python
from moabb.datasets import SSVEPExo
import matplotlib.pyplot as plt
```

```python
dataset = SSVEPExo()
X = dataset.get_data(subjects=[1])[1]['0']['0'].get_data()
X.shape
```

```python
for eletrodo in range(8):#vai rodar as oito linhas da matrix (a ultima linha foi desconsiderada pois era toda de zeros)
    plt.plot([i/256 for i in range(2*256)], 0.03*eletrodo+X[eletrodo, :2*256]) #a frequencia é 256Hz, logo a multiplicação por 2 diz que vamos plotar apenas dois segundos dos sinais
plt.yticks([])
plt.xlabel('tempo (s)')
plt.show()
```
